### Spans und Labels checken

```python
def get_label(label_id):
    label_mapping = {
        0: "Medikation",
        1: "Dosis",
        2: "Diagnose"
    }
    return label_mapping.get(label_id)

with open("/Users/wielandcremer/Documents/Uni/6_Semester/information_extraction_german_medical_data/meaningful_modification/data/synthetic/synthetic_gptnermed.jsonl", "r", encoding="utf-8") as file:
    
    for line in islice(file,5):
        line = json.loads(line)

        sentence = line["sentence"]
        ner_labels = line["ner_labels"]

        ner_class = ner_labels["ner_class"]
        start = ner_labels["start"]
        end = ner_labels["stop"]

        print(f"Sentence: {sentence}")
        for ner_class, start, end in zip(ner_class, start, end):
            entity = sentence[start:end]
            print(f"Entity: {entity}, Class: {get_label(ner_class)}")
        print("\n")        
```
