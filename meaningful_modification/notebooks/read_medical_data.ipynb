## ICD-10-GM

```python
import csv
from pathlib import Path

eingabe = Path(
    "/Users/wielandcremer/Downloads/alphaidse2026/"
    "icd10gm2026_alphaidse_edvtxt_20250926.txt"
)

ausgabe = Path(
    "/Users/wielandcremer/Dokumente/Uni/6_Semester/"
    "information_extraction_german_medical_data/"
    "meaningful_modification/data/medical_information/"
    "Alpha-ID-SE.csv"
)

ausgabe.parent.mkdir(parents=True, exist_ok=True)

anzahl = 0

with (
    eingabe.open("r", encoding="utf-8") as quelle,
    ausgabe.open("w", encoding="utf-8-sig", newline="") as ziel
):
    writer = csv.writer(
        ziel,
        delimiter=";",
        quotechar='"',
        quoting=csv.QUOTE_MINIMAL
    )

    writer.writerow(["ICD_Code", "Diagnose"])

    for zeilennummer, zeile in enumerate(quelle, start=1):
        spalten = zeile.rstrip("\r\n").split("|")

        if len(spalten) < 8:
            print(f"Zeile {zeilennummer} übersprungen")
            continue

        icd_code = spalten[5].strip()
        diagnose = spalten[7].strip()

        if icd_code and diagnose:
            writer.writerow([icd_code, diagnose])
            anzahl += 1

print(f"{anzahl} Diagnosen gespeichert unter: {ausgabe}")
```

```python
AlphaIDSE = pd.read_csv(
    ausgabe,
    sep=";",
    encoding="utf-8-sig",
    dtype=str
)
```
